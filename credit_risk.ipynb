```python
import pandas as pd
df=pd.read_csv("credit_risk.csv")
```

```python
df["person_emp_length"]=df["person_emp_length"].fillna(df["person_emp_length"].median())
df["loan_int_rate"]=df["loan_int_rate"].fillna(df["person_emp_length"].median())
```

```python
df.isna().sum().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe=OneHotEncoder(sparse_output=False)
```

```python
encoded=oe.fit_transform(df[["person_home_ownership","loan_intent","cb_person_default_on_file"]])
encoded_col=oe.get_feature_names_out(["person_home_ownership","loan_intent","cb_person_default_on_file"])
encoded_df = pd.DataFrame(encoded, columns=encoded_col, index=df.index)
df2=pd.concat([df.drop(columns=["person_home_ownership","loan_intent","cb_person_default_on_file"]),encoded_df],axis=1)
```

```python
import numpy as np
median_val = df["person_emp_length"].median()
df2["person_emp_length"] = df["person_emp_length"].apply(
    lambda x: median_val if x > 50 else x
)
```

```python
df2.head()
```

```python
df2.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df2["loan_grade_encoded"]=le.fit_transform(df["loan_grade"])
```

```python
df2.head()
```

```python
df2["age_group"]='notavail'
```

```python
df2["age_group"] = df["person_age"].apply(
    lambda x: "Young" if x < 25 else ("Adult" if x < 45 else "Senior")
)
```

```python
df2.isna().sum()
```

```python
df2=df2.drop(columns="loan_grade")
```

```python
df2.head()
```

```python
encoded=oe.fit_transform(df2[["age_group"]])
encoded_col=oe.get_feature_names_out(["age_group"])
encoded_df=pd.DataFrame(encoded,columns=encoded_col,index=df2.index)
df3=pd.concat([df2.drop(columns=["age_group"]),encoded_df],axis=1)
```

```python
df3.head()
```

```python
df3["credit_utilization_rate"]=df3["loan_amnt"]/df["person_income"]
```

```python
df3.head()
```

```python
df3["CIB"]=df3["loan_amnt"]/(df["person_income"]*df3["cb_person_cred_hist_length"])
```

```python
df3.info()
```

```python
df3.head()
```

```python
from sklearn.model_selection import train_test_split
x=df3.drop(columns=["loan_status"])
y=df3["loan_status"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
```

```python
import numpy as np
if len(y_train.shape)>1:
   y_train=np.argmax(y_train,axis=1)
if len(y_test.shape)>1:
 y_test=np.argmax(y_test,axis=1)
```

```python

ltg = lgb.LGBMClassifier(
    objective="binary",   
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1
)
```

```python
ltg.fit(x_train,y_train)
```

```python

y_pred1=ltg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```
